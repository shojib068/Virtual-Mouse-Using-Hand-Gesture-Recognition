```python
import cv2
cv2.__version__
import mediapipe as mp
import pyautogui
import random
import util

```

```python
def main():
    cap = cv2.VideoCapture(0)
    
    try:
        while cap.isOpened():
            ret, frame = cap.read()

            if not ret:
                break
            frame = cv2.flip(frame, 1)


            cv2.imshow('Frame',frame)
            if cv2.waitKey(1) & 0xFF == ord('q'): ##press q to halt
                break

    finally:
        cap.release()
        cv2.destroyAllWindows()

if __name__ == '__main__':
    main()
    
```
